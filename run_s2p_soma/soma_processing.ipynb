```python
import os

folder_path = '/path/to/folder'  # Specify the path to your folder here

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # Check if it is a file (not a directory)
    if os.path.isfile(file_path):
        # Get the parent directory name
        directory_name = os.path.basename(os.path.dirname(file_path))

        # Create the new file name by combining the directory name and original file name
        new_filename = f'{directory_name}_{filename}'

        # Rename the file
        new_file_path = os.path.join(folder_path, new_filename)
        os.rename(file_path, new_file_path)
```

```python
def export_image_files_to_suite2p_format(tiff_path):
    os.chdir(tiff_path)
    destination = tiff_path
    for file in os.listdir(tiff_path):
        if file.endswith(".tif"):

            name = file[0:-4]
            os.makedirs(name, exist_ok = True)
            destination = tiff_path + '/' + name + '/' + file
            copyfile(file, destination)
            os.remove(file)
```
